```python
import ast
def get_latitude_longitude():
    coordinates = ast.literal_eval(intersection_df["geometry"])
    latitude = coordinates['coordinates'][0][0]
    longitude = coordinates['coordinates'][0][1]
    return latitude, longitude
```

```python
intersection_df["latitude"]= 0
intersection_df["longitude"] = 0
for i in range(intersection_df.shape[0]):
    intersection_df["latitude"][i] = ast.literal_eval(intersection_df["geometry"][i])['coordinates'][0][0]
    intersection_df["longitude"][i] = ast.literal_eval(intersection_df["geometry"][i])['coordinates'][0][1]
```
