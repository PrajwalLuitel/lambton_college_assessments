# Data Visualization with Python - Week 6

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("company_sales_data.csv")
df.head()
```

### 1) The Year-end contribution of each Product, in terms of total_units_sold

```python
df_sum = pd.DataFrame()
columns = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

for column in columns:
    df_sum[column] = df[column]

df_sum.head()
```

```python
df_sum.sum()
```

```python
plt.figure(figsize=(10,10))
plt.pie(x=df_sum.sum(), autopct='%1.1f%%' , labels=columns)
plt.title("Total sales of all the products throughtout the year")
plt.show()
```

### 2)  The contribution of each Month, towards total_year-end_profit

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul","Aug", "Sep", "Oct", "Nov", "Dec"]

plt.figure(figsize=(10,10))
plt.pie(x = df['total_profit'], autopct='%1.1f%%', labels=months)
plt.title("Contribution of each Month in total sales throughtout the year")
plt.show()
```
