```python
import pandas as pd
import numpy as np

import dash
from dash import dcc
from dash import html
from dash.dependencies import Input,Output
```

```python
df = pd.read_csv("airline_data.csv")
df.head()
```

```python
pd.set_option("display.max_columns", None)
df.head()
```

```python
df["DistanceGroup"].unique()
```

```python
df2 = df["DistanceGroup"].value_counts().reset_index()
df2["Percentage"] = (df2["count"]/(df2["count"].sum()))*100
df2
```

```python
import plotly.express as px
```

```python
fig = px.pie(data_frame= df2, values="DistanceGroup", color="DistanceGroup", title="Distance group proportions by Flights", names="DistanceGroup")
fig
```

```python
# Create a dash application
app = dash.Dash(__name__)

# Create the layout of the application
app.layout = html.Div(children=[html.H1("Airline Dashboard", style={ "textAlign": "center", "padding":"1rem", "color":"#008080", "margin":"0"}),
                                html.P("Proportion of distance group (250 mile distance interval group) by flights", style={"textAlign": "center", "color":"teal", "paddng":"0.5rem", "margin":"0", "fontSize":"1.5rem", "paddingTop":"1rem","paddingBottom":"2rem"}),
                                dcc.Graph(figure=fig)],
                      style={"backgroundImage":"url('https://images.pexels.com/photos/19897324/pexels-photo-19897324/free-photo-of-aerial-view-of-rural-landscape.jpeg')", "backgroundPosition":"fixed"}
)



# Run the application
if __name__ == "__main__":
  app.run_server()
```

```python
# Create a dash application
app2 = dash.Dash(__name__)

# Create the layout of the application
app2.layout = html.Div(children=[html.H1("Airline Performance Dashboard", style={ "textAlign": "center", "padding":"1rem", "color":"#008080", "margin":"0"}),
                                html.P("The interactive chart for montly average flight delay per year", style={"textAlign": "center", "color":"teal", "padding":"0.5rem", "margin":"0", "fontSize":"1.5rem", "paddingTop":"1rem","paddingBottom":"2rem"}),
                                html.Div(["Input Year: ", dcc.Input(value =2010, type="number", id="input-year")], style={"marginBottom":"1rem", "color":"teal", "padding":"1rem"}),
                                dcc.Graph(id="line-graph")],
                      style={"backgroundImage":"url('https://images.pexels.com/photos/19897324/pexels-photo-19897324/free-photo-of-aerial-view-of-rural-landscape.jpeg')", "backgroundPosition":"fixed"}
)

# Callback decorator
@app2.callback(Output(component_id="line-graph", component_property="figure"), Input(component_id="input-year", component_property="value"))
def create_graph(input_year):
  df3 = df[df['Year']== int(input_year)]
  # Grouping the data by month and getting the average arrival delay
  graph_df = df3.groupby('Month').aggregate({'ArrDelay':'mean'}).reset_index()

  fig2 = px.line(data_frame=graph_df, x="Month", y="ArrDelay", title=f"Average monthly delay in {input_year}")
  fig2.update_traces(line=dict(color="green"))

  return fig2

# Run the application
if __name__ == "__main__":
  app2.run_server()
```
