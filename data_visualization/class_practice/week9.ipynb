```python
# How many immigrants arrived in Canada, across all years, from BRICS' countries
brics_df = df_can.set_index("Country").loc[["Brazil","India","Russian Federation","China","South Africa","Egypt","Ethiopia","United Arab Emirates"]]
brics_df
```

```python
df_can.set_index("Country", inplace=True)
```

```python
df_can.loc["Brazil"]["Total"]
```

```python
from pywaffle import Waffle
```

```python
plt.figure(FigureClass=Waffle,
           rows=40, columns = 80,
           values = brics_df["Total"],
           legend={"loc":"lower left",
                   "bbox_to_anchor":(0,-0.5),
                   "labels":[f"{k}" for k, v in zip(brics_df.index.values, brics_df["Total"])],
                   "ncol":2
                   })
plt.show()
```
