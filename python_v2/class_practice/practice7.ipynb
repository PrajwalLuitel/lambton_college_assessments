```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('boeing747.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(image, cmap="gray"), plt.axis("off")
plt.show()
```

```python
height, width = image.shape
my_image = image[height-height//2:height, width-width//2:width]

plt.imshow(my_image, cmap="gray"), plt.axis("off")
plt.show()
```

```python
image_bgr = cv2.imread('boeing747.jpg', cv2.IMREAD_COLOR)
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGBA)
```

```python
plt.imshow(image_rgb), plt.axis("off")
plt.show()
```
